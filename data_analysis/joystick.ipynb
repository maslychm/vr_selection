```python
ax = plt.figure().add_subplot(projection='3d')

X = np.arange(-1, 1.1, .1)
Y = np.arange(1, 3.1, .1)
X, Y = np.meshgrid(X, Y)
Z = np.sign(X) * X**2 * Y

colortuple = ('y', 'b')
colors = np.empty(X.shape, dtype=str)
for y in range(Y.shape[0]):
    for x in range(X.shape[0]):
        colors[y, x] = colortuple[(x + y) % len(colortuple)]

# Plot the surface with face colors taken from the array we made.

Y = Y - 1
Y = Y / 2
Y = Y * 1000

surf = ax.plot_surface(X, Y, Z, facecolors=colors, linewidth=0)

# Customize the z axis.
ax.set_zlim(-3, 3)
ax.zaxis.set_major_locator(LinearLocator(6))

ax.set_xlim(-3, 3)
ax.set_ylim(-3000, 3000)

ax.set_xlabel('JoyStick Input')
ax.set_ylabel('Time Pressed (acceleration, mS)')
ax.set_zlabel('Applied Shift (m)')

ax.set_proj_type('ortho')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

speed = 2.5

X = np.arange(-1, 1.1, .1)
Y1 = np.sign(X) * X**2 * 1 * speed
Y2 = np.sign(X) * X**2 * 3 * speed

ax.plot(X, Y1, label='Speed with no acceleration (at 0 seconds)', color="blue")
ax.plot(X, Y2, label='Speed with full acceleration (after 1 second)', color="green")
ax.legend()
ax.plot(X, Y1, "o", label='Minimum Time', color="blue")
ax.plot(X, Y2, "o", label='Maximum Time', color="green")

for i in range(len(X)):
    ax.plot([X[i], X[i]], [Y1[i], Y2[i]], linestyle="--", color="black", linewidth=.5)

ax.set_xlabel('Joystick Input Tilt')
ax.set_ylabel('Applied Shift Speed (meters per second)')

fig.tight_layout()

fig.savefig('joystick.pdf')
```
