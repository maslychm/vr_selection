```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
dfs_no_outliers_per_condition = pd.read_csv("clean.csv")
```

```python
order = [
    ["Flower", "ConeExpand"],
    ["OhMiniMap", "2DMiniMap+"],
    ["ThreeDMiniMap", "3DMiniMap"],
    ["GravityZone", "GravityZone"],
]
```

```python
ax = sns.barplot(
    data=dfs_no_outliers_per_condition, 
    x="trial_type", 
    y="trial_time", 
)
```

```python
ax = sns.barplot(
    data=dfs_no_outliers_per_condition, 
    x="trial_type", 
    y="trial_time", 
    hue="technique",
)
```

```python
ax = sns.barplot(
    data=dfs_no_outliers_per_condition, 
    x="trial_type", 
    y="trial_time", 
    hue="technique", 
    hue_order=[o[0] for o in order],
    order=["Search", "Repeat"],
    ci=95,
)

handles, labels = ax.get_legend_handles_labels()
labels = [o[1] for o in order]
# ax.legend(handles=handles[:], labels=labels[:])
ax.legend(loc="lower left", handles=handles[:], labels=labels[:])
```

```python
ax = sns.catplot(data=dfs_no_outliers_per_condition, x="trial_type", y="trial_time", col="technique", kind="box", showfliers=False)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
sns.boxplot(data=dfs_no_outliers_per_condition, x="trial_type", y="trial_time", ax=ax[0])
sns.boxplot(data=dfs_no_outliers_per_condition, x="trial_type", y="trial_time", hue="technique", ax=ax[1])
```
